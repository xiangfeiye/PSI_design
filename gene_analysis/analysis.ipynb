```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
numbins = 32
df = pd.read_csv('../candidates/candidates_pre.csv')
```

```python
new_df = pd.DataFrame()
target = 'dianhydride'
p_list = ['sp_d', 'sp_h', 'sp_p', 'td', 'fs', 'vis_pre_110']

for j in p_list:
    
    data = df[j]
    res_freq = stats.relfreq(data, numbins=numbins, defaultreallimits=[df.min()[j], df.max()[j]])
    cdf_value = np.cumsum(res_freq.frequency)
    x = res_freq.lowerlimit + np.linspace(0, res_freq.binsize * res_freq.frequency.size, res_freq.frequency.size)
    new_df['%s_x'%j] = x
    new_df['%s_y'%j] = cdf_value
    tmp_sum = cdf_value.sum()
    
    for i in range(1, 26):
        select_df = df[df[target] == i]
        data = select_df[j]
        res_freq = stats.relfreq(data, numbins=numbins, defaultreallimits=[df.min()[j], df.max()[j]])
        cdf_value = np.cumsum(res_freq.frequency)
        new_df['%s_%s'%(j, i)] = cdf_value
new_df = new_df.fillna(0)
new_df.to_csv('%s.csv'%target, index=False)

df1 = pd.read_csv('%s.csv'%target)
result_df = pd.DataFrame()
for i in range(len(p_list)):
    select_df = df1.iloc[:,i*27+2:(i+1)*27]
    mean = select_df.sum().mean()
    result = mean - select_df.sum()
    result_df['%s_%s'%(target, p_list[i])] = list(result)
    
result_df.to_csv('%s_result.csv'%target, index=False)
```
